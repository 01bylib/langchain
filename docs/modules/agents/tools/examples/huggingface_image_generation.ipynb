# HuggingFace Image Generation Tool

Agents can generate models using HuggingFace Inference Enpdoints via the HuggingFace Image Generation Tool.

To access the endpoint, you will have to copy your API Key from your [HuggingFace console](https://huggingface.co/settings/tokens).

```python
from langchain.tools.huggingface.image_generation import HuggingFaceImageGenerationTool
```

```python
api_key = "" # Copy a token with read permissions from https://huggingface.co/settings/tokens
image_generation_tool = HuggingFaceImageGenerationTool.from_api_key(api_key, "CompVis/stable-diffusion-v1-4")
```

```python
from pathlib import Path
from tempfile import TemporaryDirectory

from IPython.display import Image


out_path = "parrot.png"
image_generation_tool.run({"prompt": "a painting of a parrot playing a saxaphone", "out_path": str(out_path)})
# Show the generated image
Image(out_path)
```
